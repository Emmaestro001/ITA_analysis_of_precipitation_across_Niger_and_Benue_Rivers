```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
KAI= pd.read_excel('BOX PLOTS.xlsx', sheet_name ='KAINJI')
YEN= pd.read_excel('BOX PLOTS.xlsx', sheet_name ='YENAGOA')
MAK= pd.read_excel('BOX PLOTS.xlsx', sheet_name ='MAKURDI')
ONI= pd.read_excel('BOX PLOTS.xlsx', sheet_name ='ONITSHA')
LOK= pd.read_excel('BOX PLOTS.xlsx', sheet_name ='LOKOJA')
YOL= pd.read_excel('BOX PLOTS.xlsx', sheet_name ='YOLA')
KAI
```

```python
# Create a figure with a 2x2 grid of subplots
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(12, 15), gridspec_kw={'hspace': 0.2, 'wspace': 0.2})

# Use the first subplot for a scatter plot
sns.boxplot(data=KAI, ax=axes[0, 0])
#axes[0, 0].set_xlabel('Total Bill')
axes[0, 0].set_ylabel('Precipitation (mm)')
axes[0, 0].set_title('Kainji')

# Use the second subplot for a line plot
sns.boxplot(data=YOL, ax=axes[0, 1])
#axes[0, 1].set_xlabel('Total Bill')
axes[0, 1].set_ylabel('Precipitation (mm)')
axes[0, 1].set_title('Yola')

# Use the third subplot for a bar plot
sns.boxplot(data=MAK, ax=axes[1, 0])
#axes[1, 0].set_xlabel('Day of Week')
axes[1, 0].set_ylabel('Precipitation (mm)')
axes[1, 0].set_title('Makurdi')

# Use the fourth subplot for a histogram
sns.boxplot(data=LOK, ax=axes[1, 1])
#axes[1, 1].set_xlabel('Day of Week')
axes[1, 1].set_ylabel('Precipitation (mm)')
axes[1, 1].set_title('Lokoja')

# Use the fifth subplot for a bar plot
sns.boxplot(data=ONI, ax=axes[2, 0])
#axes[2, 0].set_xlabel('Total Bill')
axes[2, 0].set_ylabel('Precipitation (mm)')
axes[2, 0].set_title('Onitsha')

# Use the sixth subplot for a histogram
sns.boxplot(data=YEN, ax=axes[2, 1])
#axes[2, 1].set_xlabel('Party Size')
axes[2, 1].set_ylabel('Precipitation (mm)')
axes[2, 1].set_title('Yenagoa')


# Set the title of the figure
#fig.suptitle('My Seaborn Subplots')

# Display the plot
plt.show()
```

```python
import matplotlib.gridspec as gridspec

# create a figure with a grid of 2 rows and 2 columns
fig = plt.figure(figsize=(10, 10)) 
gs = gridspec.GridSpec(nrows=2, ncols=2, figure=fig, hspace=0.25, wspace=0.00)

# create Seaborn plots for each axis
ax1 = fig.add_subplot(gs[0, 0])
sns.boxplot(data=DRY, orient='h', ax=ax1)
plt.setp(ax1.get_xticklabels(), rotation=90)
ax1.set_xlabel('Precipitation (mm)')
ax1.set_title('DRY SEASON', fontweight='bold')


ax2 = fig.add_subplot(gs[0, 1])
sns.boxplot(data=WET, orient='h', ax=ax2)
plt.setp(ax2.get_xticklabels(), rotation=90)
ax2.set_xlabel('Precipitation (mm)')
ax2.set_title('WET SEASON',  fontweight='bold')
ax2.set(yticklabels=[], ylabel='')


ax3 = fig.add_subplot(gs[1, :])
sns.boxplot(data=ANN, ax=ax3)
ax3.set_title('ANNUAL',  fontweight='bold')
ax3.set_ylabel('Precipitation (mm)')


# display the plot
plt.show()
```
