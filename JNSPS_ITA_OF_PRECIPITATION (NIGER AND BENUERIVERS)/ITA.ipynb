```python
import seaborn as sns
import matplotlib.pyplot as plt

# Generate the tips dataset
import numpy as np
import pandas as pd
np.random.seed(0)
n = 500
tips = pd.DataFrame({
    'total_bill': np.random.normal(25, 5, n),
    'tip': np.random.normal(5, 1, n),
    'size': np.random.choice([1, 2, 3, 4, 5], size=n),
    'day': np.random.choice(['Thur', 'Fri', 'Sat', 'Sun'], size=n)
})

# Create a figure with a 2x2 grid of subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8), gridspec_kw={'hspace': 0.4, 'wspace': 0.4})

# Use the first subplot for a scatter plot
sns.scatterplot(x='total_bill', y='tip', data=tips, ax=axes[0, 0])
axes[0, 0].set_xlabel('Total Bill')
axes[0, 0].set_ylabel('Tip')

# Use the second subplot for a line plot
sns.lineplot(x='size', y='total_bill', data=tips, ax=axes[0, 1])
axes[0, 1].set_xlabel('Party Size')
axes[0, 1].set_ylabel('Total Bill')

# Use the third subplot for a bar plot
sns.barplot(x='day', y='total_bill', data=tips, ax=axes[1, 0])
axes[1, 0].set_xlabel('Day of Week')
axes[1, 0].set_ylabel('Total Bill')

# Use the fourth subplot for a histogram
sns.histplot(x='total_bill', data=tips, ax=axes[1, 1])
axes[1, 1].set_xlabel('Total Bill')
axes[1, 1].set_ylabel('Count')

# Set the title of the figure
fig.suptitle('My Seaborn Subplots')

# Display the plot
plt.show()
```
